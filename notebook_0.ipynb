## 1. Installation sous Windows

### Étape 1 : Télécharger Python
1. Rendez-vous sur le site officiel : https://www.python.org/downloads/
2. Cliquez sur le gros bouton jaune **"Download Python 3.x.x"** (le site détecte automatiquement que vous êtes sous Windows).
3. Un fichier d'installation (ex: `python-3.12.x-amd64.exe`) va se télécharger.

### Étape 2 : Lancer l'installation (ÉTAPE CRUCIALE)
1. Double-cliquez sur le fichier téléchargé.
2. ⚠️ **ATTENTION !** En bas de la fenêtre d'installation, cochez impérativement la case **"Add python.exe to PATH"** (Ajouter Python au PATH). C'est indispensable pour que Windows trouve Python.
3. Cliquez ensuite sur **"Install Now"**.
4. Laissez l'installation se terminer et cliquez sur "Close".

### Étape 3 : Vérifier l'installation
1. Ouvrez l'Invite de commandes : cliquez sur le menu Démarrer, tapez `cmd` et appuyez sur Entrée.
2. Tapez la commande suivante et appuyez sur Entrée :
   ```bash
   python --version
   ```
3. Si le terminal affiche `Python 3.x.x`, bravo ! Python est installé.

## 2. Installation sous Linux (Ubuntu/Debian)

Sous Linux, Python est souvent déjà installé, mais il nous faut le gestionnaire de paquets `pip` pour installer Jupyter.

### Étape 1 : Mettre à jour le système et installer pip
1. Ouvrez un **Terminal** (raccourci clavier : `Ctrl + Alt + T`).
2. Tapez les commandes suivantes une par une, en appuyant sur Entrée après chacune (votre mot de passe vous sera demandé) :
   ```bash
   sudo apt update
   sudo apt install python3 python3-pip
   ```

### Étape 2 : Vérifier l'installation
1. Dans le terminal, tapez :
   ```bash
   python3 --version
   ```
2. Si le terminal affiche `Python 3.x.x`, c'est parfait !

## 3. Installation de Jupyter Lab (Windows ET Linux)

Maintenant que Python est installé, nous allons utiliser son gestionnaire de paquets (`pip`) pour télécharger Jupyter Lab.

### Étape 1 : L'installation
Ouvrez votre terminal (l'Invite de commandes sous Windows, ou le Terminal sous Linux) et tapez exactement cette commande :

**Sous Windows :**
```bash
pip install jupyterlab
```

**Sous Linux :**
```bash
pip3 install jupyterlab
```
(Laissez l'ordinateur travailler quelques minutes, il va télécharger les fichiers nécessaires).

### Étape 2 : Lancer Jupyter Lab
Pour lancer l'interface de travail, tapez simplement dans votre terminal :

**Sous Windows :**
```bash
jupyter lab
```

**Sous Linux :**
```bash
jupyter-lab
```

👉 **Que se passe-t-il ?** Votre navigateur web par défaut (Chrome, Firefox, etc.) va s'ouvrir tout seul sur l'interface de Jupyter Lab. **Attention :** Ne fermez pas la fenêtre du Terminal pendant que vous travaillez sur Jupyter, sinon le moteur s'arrêtera !
