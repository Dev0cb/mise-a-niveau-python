## 1. Charger un dataset intégré
Seaborn vient avec des jeux de données d'exemple très utilisés en Machine Learning.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# On charge le dataset 'tips' (pourboires dans un restaurant)
tips = sns.load_dataset("tips")

# Affichons les 5 premières lignes
tips.head()
```

## 2. Scatterplot (Nuage de points)
Idéal pour voir la relation entre deux variables numériques. Ici, l'addition totale vs le pourboire, coloré par le moment de la journée.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Relation entre Addition et Pourboire")
plt.show()
```

## 3. Boxplot (Boîte à moustaches)
Parfait pour voir la distribution d'une variable et repérer les valeurs extrêmes (outliers).

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Distribution de l'addition par jour")
plt.show()
```
