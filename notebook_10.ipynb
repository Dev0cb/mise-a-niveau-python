## 1. GroupBy (Regroupement)
Le `.groupby()` permet de regrouper des lignes identiques et d'appliquer une fonction mathématique (comme la moyenne `.mean()` ou la somme `.sum()`).

```python
import pandas as pd

df = pd.DataFrame({
    'Service': ['RH', 'IT', 'RH', 'IT', 'IT'],
    'Salaire': [3000, 4000, 3200, 4500, 4200]
})

# Quelle est la moyenne des salaires par service ?
moyenne_service = df.groupby('Service')['Salaire'].mean()
moyenne_service
```

## 2. Gestion des valeurs manquantes
Dans la vraie vie, les données sont souvent incomplètes. Pandas représente ces vides par `NaN` (Not a Number). On peut les remplir avec `.fillna()` ou les supprimer avec `.dropna()`.

```python
df_manquant = pd.DataFrame({
    'A': [1, 2, None],
    'B': [4, None, 6]
})

print("DataFrame avec manquants :")
print(df_manquant)

print("\nDataFrame rempli avec des 0 :")
print(df_manquant.fillna(0))
```
